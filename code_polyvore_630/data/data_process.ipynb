```python
import pandas as pd
import numpy as np
```

```python
path = "E:/FHN/data/polyvore/tuples_630"
```

```python
u_pos = pd.read_csv(path+'/tuples_train_posi')
```

```python
u_pos
```

```python
flag=1
for u in range(630):
    pos = u_pos[u_pos['user']==u]
    pos_50 = pos.sample(n=50, axis=0)
    pos = pos.append(pos_50)
    pos = pos.append(pos_50)
    result = pos.drop_duplicates(subset=['user','top','bottom','shoe'],keep=False)
    if flag==1:
        u_pre = pos_50
        train = result
        flag=0
    else:
        u_pre = pd.concat([u_pre, pos_50])
        train = pd.concat([train, result])
```
